```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Import

```python
path = "export_files/"

generation_by_fuel = pd.read_csv(path + "generation_by_fuel.csv")
```

```python
generation_by_fuel
```

```python
fuel_to_color = {
    "biomass": "forestgreen",
    "hard coal": "black",
    "hydro": "mediumblue",
    "lignite": "saddlebrown",
    "gas": "lightskyblue",
    "uran": "gray",
    "oil": "sandybrown",
    "waste": "lime",
    "wind": "cyan",
    "sun": "yellow"
}

t = generation_by_fuel.timestep.values

y = np.vstack([list(generation_by_fuel[fuel]) 
               for fuel in list(generation_by_fuel) 
               if fuel != "timestep"]
             )

labels = [fuel for fuel in list(generation_by_fuel)
             if fuel != "timestep"
         ]

fig, ax = plt.subplots(figsize=(10,8))
ax.stackplot(t, y, labels=labels)
ax.legend()
plt.xlabel("Hourly timesteps")
plt.ylabel("MW")
plt.grid()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 1, 2, 3, 5]
y2 = [0, 4, 2, 6, 8]
y3 = [1, 3, 5, 7, 9]

y = np.vstack([y1, y2, y3])

labels = ["Fibonacci ", "Evens", "Odds"]

fig, ax = plt.subplots()
ax.stackplot(x, y, labels=labels)
ax.legend(loc='upper left')
plt.show()
```
